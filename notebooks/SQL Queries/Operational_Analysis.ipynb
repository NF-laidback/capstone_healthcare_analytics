### 1. Retrieve the top 10 departments by total visit volume.


```python
from db_utils import run_query, execute_ddl

create_view_query = """
CREATE OR REPLACE VIEW vw_operational_department_visit_volume AS
SELECT
    department,
    COUNT(*) AS total_visits
FROM visits
GROUP BY department
ORDER BY total_visits DESC
LIMIT 10;
"""

execute_ddl(create_view_query)

# Pull the view's data into a DataFrame to verify
df_dept_volume = run_query("SELECT * FROM vw_operational_department_visit_volume;")
df_dept_volume
```

### 2. Identify the top 5 departments with the highest average length of stay.


```python
create_view_query_avg_stay_length = """
CREATE OR REPLACE VIEW vw_operational_department_avg_stay_length AS
SELECT department,
AVG(length_of_stay_hours) AS avg_length_of_stay_hrs
FROM visits
GROUP BY department
ORDER BY avg_length_of_stay_hrs DESC
LIMIT 5;
"""

execute_ddl(create_view_query_avg_stay_length)

# Pull the view's data into a DataFrame to verify
df_dept_avg_stay_length = run_query("SELECT * FROM vw_operational_department_avg_stay_length;")
df_dept_avg_stay_length
```

### 3 Find the percentage of High Risk visits per department.


```python
create_view_query_high_risk_pct = """
CREATE OR REPLACE VIEW vw_operational_department_high_risk_pct AS
SELECT department,
ROUND(SUM(risk_score = "High")/ COUNT(risk_score)* 100,2) as percentage_high_risk_visits
FROM visits
GROUP BY department
ORDER BY percentage_high_risk_visits desc;
"""

execute_ddl(create_view_query_high_risk_pct)

df_high_risk_pct = run_query("SELECT * FROM vw_operational_department_high_risk_pct;")
df_high_risk_pct
```

### 4 Determine the average number of visits per patient by city.


```python
create_view_query_visits_per_patient = """
CREATE OR REPLACE VIEW vw_operational_city_avg_visits_per_patient AS
SELECT city,
ROUND(COUNT(visit_id) / COUNT(DISTINCT visits.patient_id), 2) as avg_visits_per_patient
FROM visits JOIN patients
ON visits.patient_id = patients.patient_id
GROUP BY city
ORDER BY avg_visits_per_patient DESC;
"""

execute_ddl(create_view_query_visits_per_patient)

df_visits_per_patient = run_query("SELECT * FROM vw_operational_city_avg_visits_per_patient;")
df_visits_per_patient
```

### 5 Identify doctors handling the highest number of High Risk visits.


```python
create_view_query_doctor_high_risk = """
CREATE OR REPLACE VIEW vw_operational_doctor_high_risk_visits AS
SELECT doctor_id,
SUM(risk_score = "High") as high_risk_visit_count
FROM visits
GROUP BY doctor_id
ORDER BY high_risk_visit_count DESC;
"""

execute_ddl(create_view_query_doctor_high_risk)

df_doctor_high_risk = run_query("SELECT * FROM vw_operational_doctor_high_risk_visits;")
df_doctor_high_risk
```

# FINANCIAL ANALYSIS

### 1 Retrieve the top 10 insurance providers by total billed amount.


```python
create_view_query_insurance_total_billed = """
CREATE OR REPLACE VIEW vw_financial_insurance_provider_total_billed AS
SELECT insurance_provider,
ROUND(SUM(billed_amount),2) as total_billed_amount
FROM visits as t1 JOIN billing as t2 
ON t1.visit_id = t2.visit_id
JOIN patients as t3 
ON t1.patient_id = t3.patient_id
GROUP BY insurance_provider
ORDER BY total_billed_amount DESC
LIMIT 10;
"""

execute_ddl(create_view_query_insurance_total_billed)

df_insurance_total_billed = run_query("SELECT * FROM vw_financial_insurance_provider_total_billed;")
df_insurance_total_billed
```

### 2. Identify the top 5 insurance providers with the highest claim rejection rate.


```python
create_view_query_claim_rejection_rate = """
CREATE OR REPLACE VIEW vw_financial_insurance_claim_rejection_rate AS
SELECT insurance_provider,
ROUND(SUM(t2.claim_status = "Rejected") / COUNT(t2.claim_status) * 100, 2) as claim_rejection_rate
FROM visits as t1 JOIN billing as t2 
ON t1.visit_id = t2.visit_id
JOIN patients as t3 
ON t1.patient_id = t3.patient_id
GROUP BY insurance_provider
ORDER BY claim_rejection_rate DESC
LIMIT 5;
"""

execute_ddl(create_view_query_claim_rejection_rate)

df_claim_rejection_rate = run_query("SELECT * FROM vw_financial_insurance_claim_rejection_rate;")
df_claim_rejection_rate
```

### 3 Find the average payment delay (payment_days) by insurance provider.


```python
create_view_query_payment_delay = """
CREATE OR REPLACE VIEW vw_financial_insurance_payment_delay_claim_status_wise AS
SELECT insurance_provider,
claim_status,
ROUND(AVG(t2.payment_days), 2) as avg_payment_delay
FROM visits as t1 JOIN billing as t2 
ON t1.visit_id = t2.visit_id
JOIN patients as t3 
ON t1.patient_id = t3.patient_id
GROUP BY insurance_provider, claim_status
ORDER BY insurance_provider, claim_status;
"""

execute_ddl(create_view_query_payment_delay)

df_payment_delay = run_query("SELECT * FROM vw_financial_insurance_payment_delay_claim_status_wise;")
df_payment_delay
```

### 4 Calculate the revenue realization ratio (approved_amount / billed_amount) by department.


```python
create_view_query_revenue_realisation = """
CREATE OR REPLACE VIEW vw_financial_department_revenue_realisation_ratio AS
SELECT department,
ROUND(SUM(approved_amount)/SUM(billed_amount), 2) as revenue_realisation_ratio
FROM visits as t1 JOIN billing as t2
ON t1.visit_id = t2.visit_id 
GROUP BY t1.department
ORDER BY revenue_realisation_ratio DESC;
"""

execute_ddl(create_view_query_revenue_realisation)

df_revenue_realisation = run_query("SELECT * FROM vw_financial_department_revenue_realisation_ratio;")
df_revenue_realisation
```

### 5. Identify visits where billed_amount is high but approved_amount is zero or missing.


```python
create_view_query_high_billed_zero_approved = """
CREATE OR REPLACE VIEW vw_financial_high_billed_zero_approved AS
SELECT 
    t1.visit_id,
    t1.department,
    t2.billed_amount,
    t2.approved_amount,
    t2.claim_status
FROM visits as t1 JOIN billing as t2
ON t1.visit_id = t2.visit_id
WHERE t2.billed_amount > 20000
AND (t2.approved_amount = 0 OR t2.approved_amount IS NULL)
ORDER BY t2.billed_amount DESC;
"""

execute_ddl(create_view_query_high_billed_zero_approved)

df_high_billed_zero_approved = run_query("SELECT * FROM vw_financial_high_billed_zero_approved;")
df_high_billed_zero_approved
```

# Data Quality and Integrity Checks


### 1. Detect visits that do not have a corresponding billing record.


```python
create_view_query_no_billing_record = """
CREATE OR REPLACE VIEW vw_data_quality_visits_no_billing_record AS
SELECT t1.visit_id
FROM visits as t1 LEFT JOIN billing as t2
ON t1.visit_id = t2.visit_id
WHERE t2.bill_id IS NULL;
"""

execute_ddl(create_view_query_no_billing_record)

df_no_billing_record = run_query("SELECT * FROM vw_data_quality_visits_no_billing_record;")
df_no_billing_record
```

### 2. Detect billing records that do not have a corresponding visit.


```python
create_view_query_no_visit_record = """
CREATE OR REPLACE VIEW vw_data_quality_billing_no_visit_record AS
SELECT t1.bill_id
FROM billing as t1 LEFT JOIN visits as t2
ON t1.visit_id = t2.visit_id
WHERE t2.visit_id IS NULL;
"""

execute_ddl(create_view_query_no_visit_record)

df_no_visit_record = run_query("SELECT * FROM vw_data_quality_billing_no_visit_record;")
df_no_visit_record
```

### 3. Identify patients with duplicate patient_id values.


```python
create_view_query_duplicate_patients = """
CREATE OR REPLACE VIEW vw_data_quality_duplicate_patient_ids AS
SELECT patient_id,
COUNT(*) as occurrences
FROM patients
GROUP BY patient_id
HAVING COUNT(*) > 1;
"""

execute_ddl(create_view_query_duplicate_patients)

df_duplicate_patients = run_query("SELECT * FROM vw_data_quality_duplicate_patient_ids;")
df_duplicate_patients
```

### 4. Find records with missing or invalid length_of_stay_hours or payment_days.


```python
create_view_query_missing_invalid_fields = """
CREATE OR REPLACE VIEW vw_data_quality_missing_invalid_stay_payment AS
SELECT 
    t1.visit_id,
    t1.length_of_stay_hours,
    t2.payment_days
FROM visits as t1 JOIN billing as t2
ON t1.visit_id = t2.visit_id
WHERE t1.length_of_stay_hours IS NULL
    OR t1.length_of_stay_hours <= 0
    OR t2.payment_days IS NULL
    OR t2.payment_days <= 0;
"""

execute_ddl(create_view_query_missing_invalid_fields)

df_missing_invalid = run_query("SELECT * FROM vw_data_quality_missing_invalid_stay_payment;")
df_missing_invalid
```

### 5. Identify visits linked to patients with missing insurance provider information.


```python
create_view_query_missing_insurance = """
CREATE OR REPLACE VIEW vw_data_quality_visits_missing_insurance AS
SELECT
    t1.visit_id,
    t1.patient_id,
    t1.visit_date,
    t1.department,
    t2.insurance_provider
FROM visits as t1 JOIN patients as t2
ON t1.patient_id = t2.patient_id
WHERE t2.insurance_provider IS NULL
    OR t2.insurance_provider = '';
"""

execute_ddl(create_view_query_missing_insurance)

df_missing_insurance = run_query("SELECT * FROM vw_data_quality_visits_missing_insurance;")
df_missing_insurance
```
